## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data=pd.merge(mouse_metadata,study_results, on="Mouse ID")
combined_data.dtypes
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
new_data=combined_data.drop_duplicates(subset=["Mouse ID", "Timepoint"])
```

```python
# Checking the number of mice in the clean DataFrame.
new_data["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for 
#each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.
```

```python
DR_mean=new_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
```

```python
DR_median=new_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
```

```python
DR_var=new_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
```

```python
DR_std=new_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
```

```python
DR_sem=new_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()
```

```python
DR_summary=pd.DataFrame({
    "Tumor Volume mean":DR_mean,
    "Tumor Volume median":DR_median,
    "Tumor Volume variance":DR_var,
    "Tumor Volume Std. Deviation":DR_std,
    "Tumor Volume SEM":DR_sem
})
DR_summary
```

## Bar Plots

```python
# Generate a bar plot showing the number of mice per time point for each treatment 
#throughout the course of the study using pandas. 
TP_count=pd.DataFrame(new_data.groupby(["Drug Regimen","Timepoint"])["Mouse ID"].count())
DR=new_data["Drug Regimen"].unique()
for x in DR:
    plot=TP_count.loc[x]
    plot.plot(kind="bar")
    plt.title(f"Number of Mice per Time point in {x} treatment")
    plt.ylabel("Number of mice")
    plt.show()
    
        
        
    
```

```python
# Generate a bar plot showing the number of mice per time 
#point for each treatment throughout the course of the study using pyplot.
for x in DR:
    xaxis=np.arange(len(TP_count.loc[x]["Mouse ID"]))
    yaxis=TP_count.loc[x]["Mouse ID"].tolist()
    plt.bar(xaxis,yaxis)
    plt.title(f"Number of Mice per Time point in {x} treatment")
    plt.xticks(xaxis,DR, rotation=45)
    plt.ylabel("Number of Mice")
    plt.show()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. 
#Calculate the IQR and quantitatively determine if there are any potential outliers. 
DR_TV=pd.DataFrame(new_data[new_data["Timepoint"]==45].groupby(["Drug Regimen","Mouse ID"])["Tumor Volume (mm3)"].unique())
sort=DR_summary.sort_values(by="Tumor Volume mean", ascending=True)
mptr=sort.index.unique().tolist()
mptr_4=[]
for i in [0,1,2,3]:
    mptr_4.append(mptr[i])
mptr_4
```

```python
for x in mptr_4:
    DF=DR_TV.loc[x]["Tumor Volume (mm3)"].astype(int)
    quantiles=DF.quantile([.25,.5,.75])
    lowerq=quantiles[.25]
    upperq=quantiles[.75]
    iqr=upperq-lowerq
    lower_bound=lowerq-(1.5*iqr)
    upper_bound=upperq+(1.5*iqr)
    print(f"For treatment regimen {x} values below {lower_bound} and above {upper_bound} could be outliers.")
    fig1, ax1 = plt.subplots()
    ax1.set_title('Treatment Regimens')
    ax1.set_ylabel('Tumor Volume')
    ax1.boxplot(DF)
    plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
line_plot_data=new_data[new_data["Drug Regimen"]=="Capomulin"]
line_plot_data.groupby("Timepoint")["Tumor Volume (mm3)"].mean().plot(label="Capomulin",title="Tumor Volume vs Timepoint")
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
scatter_data=pd.DataFrame(line_plot_data.groupby("Weight (g)")["Tumor Volume (mm3)"].mean())
scatter_data=scatter_data.reset_index()
scatter_data.plot(kind="scatter",x="Weight (g)",y="Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
x_values = scatter_data["Weight (g)"]
y_values = scatter_data["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(22,38),fontsize=15,color="red")
plt.xlabel('Weight')
plt.ylabel("Average Tumor Volume")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
